Запуск `experiments_cp2` и просмотр итоговой таблицы экспериментов.

```python
from __future__ import annotations

from pathlib import Path
import os
import subprocess
import sys
import pandas as pd

ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd()
PY = sys.executable
os.environ["CP2_FAST"] = os.environ.get("CP2_FAST", "1")

def run(cmd: list[str]) -> None:
    print("$", " ".join(cmd))
    subprocess.run(cmd, cwd=ROOT, check=True)
```

```python
run([PY, "-m", "src.modeling.experiments_cp2", "--no-mlflow"])
```

```python
exp_path = ROOT / "report" / "tables" / "experiments_cp2.csv"
df = pd.read_csv(exp_path)
print("rows:", len(df), "| path:", exp_path)
df.sort_values("val_roc_auc", ascending=False).head(10)
```
