# 07 — CP2: dim reduction + final model

Запуск `dim_reduction`, `final_model` и `time_split_eval` с просмотром основных артефактов.

```python
from __future__ import annotations

from pathlib import Path
import os
import subprocess
import sys
import pandas as pd

ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd()
PY = sys.executable
os.environ["CP2_FAST"] = os.environ.get("CP2_FAST", "1")

def run(cmd: list[str]) -> None:
    print("$", " ".join(cmd))
    subprocess.run(cmd, cwd=ROOT, check=True)
```

```python
run([PY, "-m", "src.modeling.dim_reduction"])
run([PY, "-m", "src.modeling.final_model", "--fast"])
run([PY, "-m", "src.modeling.time_split_eval"])
```

```python
final_metrics = ROOT / "report" / "tables" / "final_metrics.csv"
time_metrics = ROOT / "report" / "tables" / "time_split_metrics.csv"
perm = ROOT / "report" / "tables" / "permutation_importance.csv"

print("final_metrics:", final_metrics)
display(pd.read_csv(final_metrics))
print("time_split_metrics:", time_metrics)
display(pd.read_csv(time_metrics))
print("top permutation importance:")
display(pd.read_csv(perm).head(15))
```
