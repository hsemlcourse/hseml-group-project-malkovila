# 05 — CP2: features, time split, drift

Этот ноутбук запускает шаги подготовки данных CP2 и показывает ключевые артефакты.

```python
from __future__ import annotations

from pathlib import Path
import subprocess
import sys
import pandas as pd

ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd()
PY = sys.executable

def run(cmd: list[str]) -> None:
    print("$", " ".join(cmd))
    subprocess.run(cmd, cwd=ROOT, check=True)
```

```python
run([PY, "-m", "src.features.build_dataset", "--full"])
run([PY, "-m", "src.features.time_split"])
run([PY, "-m", "src.features.drift_report"])
```

```python
meta_path = ROOT / "data" / "processed" / "split_meta.json"
feature_drift = ROOT / "report" / "tables" / "feature_drift.csv"

print("split_meta:", meta_path)
print(meta_path.read_text(encoding="utf-8")[:500], "...\n")
print("feature_drift:", feature_drift)
pd.read_csv(feature_drift).head(10)
```
